- In order to get name and set name **directly**, def get_name and def set_name are defined as followed.

````python
class Person(object):
    
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
    
    def get_name(self):
        return self.__name
    
    def get_age(self):
        if self.__name == "zhangsan":     # the age excpet zhangsan is inaccessible.
            return self.__age              
        else:
            return "NO"
        
    def set_name(self, value):
        self.__name = value
    
    def set_age(self, value):
        self.__age = value
    
    name = property(get_name, set_name)   # enable the access and the assignment facility of name by property func.
    age = property(get_age, set_age)      # However, these expressions are expected to get rid of for simplicity.
        
p0 = Person("zhangsan", 40)
p1 = Person("lisi", 50)
p0.age = 18                               # attr assignment is done due to property func.
print(p0.name, p0.age)                    # attr access is done due to property func.
print(p1.name, p1.age)
p1.name = "awge"
````

- <code>age = property(get_age, set_age)</code><br><br>However, these expression above is expected to get rid of for simplicity. Therefore, @property is introduced as followed.

### best solution

````python
class Person(object):
    
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
    '''
    Give the set_name and set_name same function name. 
    With a decorator attached to them, name can be accessed or set easilly.
    '''
    @property
    def name(self):                    
        return self.__name
    
    @name.setter
    def name(self, value):
        self.__name = value
    
    @property
    def age(self):
        if self.__name == "zhangsan":
            return self.__age
        else:
            return "NO"
    
    @age.setter
    def age(self, value):
        self.__age = value
        
p0 = Person("zhangsan", 40)
p1 = Person("lisi", 50)
p0.age = 18
print(p0.name, p0.age)
print(p1.name, p1.age)
````

## Inheritance

In the following code, self is always bb even when the program is in the frame of AA. Abstractly speaking, self at firstly utilizes its own attributes then its father. 

````python
class AA():
    def __init__(self):
        self.n = 10
    def add(self,m):
        print(" this is AA 's add")
        self.n += m
class BB(AA):
    def __init__(self):
        self.n = 20
    def add(self,m):
        print(" this is BB 's add")
        super().add(m)
        self.n += m

bb = BB()
bb.add(100)
print(bb.n)
````

- **\[takeaway\]** self.method() will look for method() in the current class first, then follow the list of inherited classes from left to right until the method is found. The super().method() will do the same, except that it skips the current class.


````python
class A:
    def __init__(self):
        self.n = 2

    def add(self, m):
        print('self is {0} @A.add'.format(self))
        self.n += m
        print(f"self.n = {self.n}")
class B(A):
    def __init__(self):
        self.n = 3

    def add(self, m):
        print('self is {0} @B.add'.format(self))
        super().add(m)
        self.n += 3
        print(f"self.n = {self.n}")
class C(A):
    def __init__(self):
        self.n = 4

    def add(self, m):
        print('self is {0} @C.add'.format(self))
        super().add(m)
        self.n += 4
        print(f"self.n = {self.n}")
class D(B, C):
    def __init__(self):
        self.n = 5

    def add(self, m):
        print('self is {0} @D.add'.format(self))
        super().add(m)
        self.n += 5
        print(f"self.n = {self.n}")
d = D()
d.add(2)
print(d.n)
````

The above input means compute the following equation given $ x = 2, y = 3 $.<br>

$$ \frac {\partial \frac {2x - 6y}{3x + 4y}}  {\partial x} $$

````python
class Exp:                        
    def eval(self, **value):  
        '''
        eval is a method where input if any is an arg dict, which will be plugged into the 
        corresponding place of the expression. If no place is ready for a given input or unknown var remains unplugged in, error will
        be raised.
        
        The args dict will be parsed and evaluated as an expression down to a number.
        
        e.g.  [(x+2y)/3].eval(x = 1, y = 2)
        Here, x = 1 goes to x, y = 2 goes to y
        
        e.g.  [(x+2y)/3].eval(x = 1, y = 2, z = 3)
        There is no place for z in the expression, which raises an error.
        '''
        pass
    
    def deriv(self, x):           # compute the derivative wrt a given x
        pass
    
    def __add__(self, other):     # two Exp_obj can be added
        return Add(self, other)   # just an interface. The implementation of Add is coming in the next cell.
        
    def __sub__(self, other):     # two Exp_obj can be substracted from each other.
        return Sub(self, other)   
    
    def __mul__(self, other):
        return Mul(self, other)   
    
    def __truediv__(self, other): # the true divisoin always produces a true and floating point result.
        return TrueDiv(self, other)
    
````

````python
class Const(Exp):
    def __init__(self, value):
        self.value = value
        
    def eval(self, **values):     # get the value of the constant
        return self.value         # regardless of the input, eval a const remains the const itself.
    
    def deriv(self, x):           # take the derivative of the constant regardless of the input.
        return Const(0)           # don't return 0 directly, which is explained in question 1 in the next part.
    
    def __repr__(self):           # the representation of a const should be its value
        return f"{self.value}"
````

````python
class Variable(Exp):
    def __init__(self, name):            # keep the name of var the same as its name attribute. That is the_name = Variable("the_name") 
        self.name = name
        
    def eval(self, **values):            #  Here output the value of the Variable e.g. x.eval(x = 1)->correct  x.eval(y = 1)->error
        if self.name in values:          #  if the name of the input arg(e.g. x = 1) matches the name of Variable when created
            return values[self.name]     #  then output the value of x.
        raise Exception(f"Variable {self.name} is not found")
        
    def deriv(self, x:str):              #  Obj: compute dx/dx 
        if not isinstance(x, str):
            raise Exception(f' please input a string like "x" ')
        return Const(1 if x == self.name else 0)
    
    def __repr__(self):
        return f"{self.name}"
````

- So far, we have been able to create any constant obj with a **value** and variable obj with a **name**. Not only can we **evaluate** them by eval method (where for const, it's value, and for var, it's name), but also **take the derivative** of them by deriv method (where for const, it's 0, for var, it's 1)

- Now, it should have been the time to make our expression with several var and const. But don't forget about we need four types of operations to connect our var and const (recall the interfaces in class Exp). Let's create classes for them.

````python
class Add(Exp):
    def __init__(self, left, right):
        self.left = left
        self.right = right
    
    def eval(self, **values):
        return self.left.eval(**values) + self.right.eval(**values)
    
    def deriv(self, x):
        """
        (f(x)+g(x))' = f'(x) + g'(x)  f(x) and g(x) are exp, var or const
        """
        return self.left.deriv(x) + self.right.deriv(x)
    
    def __repr__(self):
        return f"({self.left} + {self.right})"
````

````python
# unit test
c1 = Const(10)
c2 = Const(20)

print((c1 + c2).eval())
print((c1 + c2).deriv("x").eval())

x = Variable("x")
y = Variable("y")

print((x+y).eval(x=2, y=4))
print((x+y).deriv("y").eval())
````

- Same philosopy apples to class Sub, Mul, TrueDiv. The only difference is def deriv.

````python
class Sub(Exp):
    def __init__(self, left, right):
        self.left = left
        self.right = right
    
    def eval(self, **values):
        return self.left.eval(**values) - self.right.eval(**values)
    
    def deriv(self, x):
        return self.left.deriv(x) - self.right.deriv(x)
    
    def __repr__(self):
        return f"({self.left} - {self.right})"
````

````python
class Mul(Exp):
    def __init__(self, left, right):
        self.left = left
        self.right = right
    
    def eval(self, **values):
        return self.left.eval(**values) * self.right.eval(**values)
    
    def deriv(self, x):  #  (uv)' = u'v + uv'
        u, v = self.left, self.right
        return u.deriv(x) * v + u * v.deriv(x)
    
    def __repr__(self):
        return f"({self.left} * {self.right})"
````

````python
class TrueDiv(Exp):
    
    def __init__(self, left, right):
        self.left = left
        self.right = right
    
    def eval(self, **values):
        return  self.left.eval(**values) / self.right.eval(**values)
    
    def deriv(self, x):  #  (u/v)' = (u'v - uv')/v**2
        u, v = self.left, self.right
        return (u.deriv(x) * v - u * v.deriv(x))/v*v
    
    def __repr__(self):
        return f"{self.left} / {self.right}"
````

````python
# unit test
c1 = Const(2)
c2 = Const(6)
c3 = Const(3)
c4 = Const(4)

x = Variable("x")
y = Variable("y")

((c1 * x - c2 * y) / (c3 * x + c4 * y)).deriv("x").eval(x=2, y=3)
````
