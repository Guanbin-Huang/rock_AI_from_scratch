```python
def insertionSort(lista):
    sorted_list = []
    sorted_list.append(lista.pop(0))  # pop out the 0-th element from lista into sorted_list

    while len(lista) != 0:            # while loop until there is no element left in the lista since we put elements into sorted_list one by one.
        ele = lista.pop(0)            # arrange for the next element into sorted_list
        sorted_list.insert(0, ele)    # insert it to the head of the sorted_list

        for i in range(len(sorted_list)-1):
            if  sorted_list[i] > sorted_list[i + 1]:  # by now, we deem that all ele have been in the right order except the 0-th ele. It's time to consider where it should go and settle.
                sorted_list[i], sorted_list[i + 1] = sorted_list[i + 1], sorted_list[i]  # the ele poped in does compare-and-swap down to the last ele in the sorted_list if no bigger is found. Otherwise, just stop where it is.
            

    return sorted_list

lista = [53, 1, 27, 36, 15, 69, 42]
sorted_lista = insertionSort(lista)
sorted_lista
```

## Shuttle Sort

**Reference**: https://www.youtube.com/watch?v=h1-_kHnfjUg<br>
The code below is derived from the above youtube video.

```python
def shuttleSort(lista):
    for i in range(len(lista) - 1):  #  how many passes are needed is based on how many elements is in the list.
        if lista[i] < lista[i+1]:    # if the pass is going well, no swap is required.
            continue                 # then jump to next pass.
            
        for i_bw in range(i,-1,-1):  # if the pass is not going well, shuttle run is needed.
            if lista[i_bw] > lista[i_bw + 1]:  # if the former is bigger than the latter, a swap occurs.
                lista[i_bw + 1], lista[i_bw] = lista[i_bw], lista[i_bw + 1]
            else:                              
                break
            
    return lista

lista = [10, 2, 7, 3, 5]
shuttleSort(lista)
lista
```

## Shell sort

**Reference:** https://www.youtube.com/watch?v=2_DbZjInBA8

```python
def shellSort(lista): 
    '''
    gap_size = 4 looks like as below.
    [4, 8, 9, 10||, 7, 1, 9, 12||, 3, 6]
    we want to take every 4 elements from the lista and see them as sublists.
    In details,
    4 7 3 are the 1-st sublist; 8 1 6 is another sublist(2nd sublist);
    9 9 (3rd sublist); 10 12 (4th sublist).
    '''
    n = len(lista)
    gap_size = int(n//2)
    
    while(gap_size>=1):             # the num of element in the sublist is 1 when gap_size = 1.
        for i in range(gap_size):   # it amounts to the num of sublists.
            sorted_sublist = insertionSort(lista[i::gap_size])  # syntax: lista[i::gap_size] means taking elements with gap_size = 4 as an i-th sublist
            lista[i::gap_size] = sorted_sublist               # put the elements to the corresponding place in the lista with a distance apart within themselves. 

        gap_size = int(gap_size//2) # reduce the gap_size until 1.
       
    return lista

lista = [4, 8, 9, 10, 7, 1, 9, 12, 3, 6]

sorted_lista = shellSort(lista)
sorted_lista
```

# Attribute and Method of Class and Instance | Static Method


- 什么是: 类,实例?
- 什么是类对象, 实例对象?
- 什么是: 类属性,类方法, 实例属性, 实例方法?


先看以下代码:

```python
class C():
    num1 = 10          # 类属性
    def __init__(self):
        self.num2 = 20 # 实例属性
    
```

上面的代码中, num1 是类属性, 而 num2 是实例属性
<br><br>

```python
obj1 = C
obj2 = C()
print(f"obj1's type: {(obj1)}")
print(f"obj2's type: {(obj2)}")
```

显然 变量 obj1 是类名,这里认为是类对象, 变量 obj2 是 类名() 的形式, 属于实例对象

**[takeaway]**: <br>
A concrete obj can access everything of its abstraction. On the other hand it is not.


**details:**<br>Instance can access instance attr and class attr<br>
Class cannot access instance attr but class attr



# Decorator

What is the right output of the following code?

```python
def debug(func):
    print(2)
    def wrapper():
        print(4)
        print(f"debug {func.__name__} haha ")
        return func()
    print(5)
    return wrapper
print(3)
@debug           # @debug can be rewritten as say_hello = debug(say_hello)
def say_hello():
    print(6)

print(1)
s = say_hello()  # the say_hello here is the wrapper (a.k.a improved_say_hello())
print(7)
```
