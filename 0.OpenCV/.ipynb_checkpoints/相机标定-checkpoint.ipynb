# 目标
- 在这次的教程当中，我们会学到下面的东西：
    - 相机畸变的类型
    - 如何找到相机的内参和外参
    - 基于这些参数，我们如何去除畸变

# 基础知识
- 针孔相机虽然是完美模型，但是进光量过少，因此科学家们引入了透镜去聚集光，但是由于制造工艺，透镜会使图像产生明显的失真。 
- 两种主要的畸变是径向畸变和切向畸变。径向畸变使直线出现弯曲。 距离图像中心越远的点，径向畸变越大。 例如，下图显示了一个棋盘的两条边用红线标出的图像。 但是，你可以看到棋盘的边界不是一条直线。 直线变凸了。

- 径向畸变可以由下面的公式来表示 
    - $x_{distorted} = x (1 + k_1r^2 + k_2r^4 + k_3r^6)$
    - $y_{distorted} = y (1 + k_1r^2 + k_2r^4 + k_3r^6)$
    - 注意这里的xy不是，离散的像素坐标，而是连续的2d图像坐标

- 类似的，切向畸变的发生是由于透镜与成像平面有些不平行。
    - $x_{distorted} = x + [2 p_1 x y  + p_2 (r^2 + 2x^2)]$
    - $y_{distorted} = y + [p_1(r^2 + 2y^2) + 2p_2xy]$

- 简而言之，我们需要5个参数，他们也叫做畸变参数
    - distortion coeffients =  $(k_1, k_2, p_1, p_2, k_3)$
 

- 除此之外，我们还需要其他的一些信息，比如说相机的内参和外参，相机的内参因相机的不同而不同。内参包括焦距的长度$(f_x, f_y)$ 和 焦点的位置$(c_x, c_y)$，这些信息可以用来构造相机矩阵（Camera matrix）

## 内参
- 相机矩阵可以用来去除畸变影响，因为相机矩阵对于特定相机来说是唯一的，即，一旦相机矩阵确定下来了，它就可以复用来拍摄并矫正其他图片的畸变。
    - camera matirx = $
\left[
\begin{array}{rcl}
f_x & 0   & c_x \\
0   & f_y & c_y \\
0   & 0   & 1   
\end{array}
\right]
$

## 外参
- 外部参数指的是旋转矩阵和平移向量（有些数学表达将平移也融入进了旋转矩阵，成为旋转平移矩阵）
- 旋转平移矩阵A坐标系转换到B坐标系（在我们的案例中，我们一般指的是将工件坐标系转到相机坐标系）
