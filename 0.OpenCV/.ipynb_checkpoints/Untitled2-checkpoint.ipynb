```python
import numpy as np
import struct
import random
import matplotlib.pyplot as plt
import pandas as pd
import math
import cv2
import math
```

```python
def crop_resize_to(src, roi, dst):
    # -------------- ROI cropping ------------------
    x, y, r, b = roi
    roi_img = src[y: b+1, x: r+1, ...]

    #     w = r - x + 1  计算长度的时候我们一般习惯于 末尾像素-头像素 +1
    #     center = (x + r) // 2
    rh, rw = roi_img.shape[:2]
    dst_h, dst_w = dst.shape[:2]
    
    # -------------- Transformation M ------------------
    scale = min(dst_h / rh, dst_w / rw) # ！！分母要是dst, 分子是src（这里roi就是src）
    t_w = -rw * 0.5 * scale + dst_w * 0.5 # 加号前是红色的平移 和 缩放 ，加号后是蓝色的平移
    t_h = -rh * 0.5 * scale + dst_h * 0.5
    
    M = np.array([[scale, 0, t_w],                        # 缩放并平移的变换矩阵，scale为缩放系数，t_w和t_h为平移的参数
                    [0, scale, t_h]])
    # -------------- WarpAffine ------------------
    
    cv2.warpAffine(roi_img, M, (dst_w, dst_h), dst = dst)
    
```

```python
img = cv2.imread("lqd.jpg")
plt.subplot(1, 2, 1)                                      # 1行2列中的第一个子图
plt.imshow(img[...,::-1]) 

dst = np.zeros((200, 200, 3), dtype = np.uint8)
crop_resize_to(img, roi = (63, 138, 203, 186), dst = dst) # xyrb

print(dst.dtype)
print(np.max(dst))

plt.subplot(1, 2, 2)
plt.imshow(dst[...,::-1])
```
